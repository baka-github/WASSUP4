## quiz 2 : 네이버 증권 주요 뉴스의 제목과 URL을 모두 추출하여 출력하라.

#### 제목 변수, url 변수 만들기

```python
# 제목과 url 데이터 목록 만들기
news_title = []
news_url = []

for item in srclist:
    title = item.text
    url = urljoin(PATH, item['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)   
```

```python
news_title
```

```python
news_url
```

#### 데이터프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame( {'제목':news_title, '주소':news_url})
df
```

#### excel 파일 저장

```python
df.to_excel('output/naver_finance.xlsx', index=False)
```

#### 파일명에다가 저장 날짜 추가하기

```python
# 파일명에 저장 시각을 추가하기
import time
```

```python
today = time.localtime()
today
```

```python
df.to_excel(f'output/{today.tm_year}_{today.tm_mon}_{today.tm_mday}_naver_finance.xlsx', index = False)
```
