#### 02-01_Pandas 패키지 소개 연습 문제

다음 데이터프레임에서 지정하는 데이터를 뽑아내거나 처리하라.

```
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
```

(1) 모든 학생의 수학 점수를 시리즈로 나타낸다.

(2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.

(3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.

(4) 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.

(5) 춘향의 점수를 데이터프레임으로 나타낸다.

(6) 향단의 점수를 시리즈로 나타낸다.

````python
import pandas as pd
````

````python
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)
df
````

````python
# (1) 모든 학생의 수학 점수를 시리즈로 나타낸다.
df['수학']
````

````python
# (2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
df[['국어', '영어']]
````

````python
# (3) 모든 학생의 모든 과목의 평균 점수를 새로운 열로 추가한다.
df.mean(axis=0)
````

````python
df.mean(axis=1)
````

````python
df["평균 점수"] = round(df.mean(axis=1))
df
````

````python
# (4) 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.
df['영어']['방자']
````

````python
df['영어']['방자'] = 80
df["평균 점수"] = round(df.mean(axis=1))
df
````

````python
# (5) 춘향의 점수를 데이터프레임으로 나타낸다.
df[:1]
````

````python
# (6) 향단의 점수를 시리즈로 나타낸다.
df.T['향단']
````

`reset_index` 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 `df_score2`을 만든다.
